# Q1. Divisible by 7, not by 5

```python
def find_numbers(lower, upper):
    lst = []
    for val in range(lower,upper+1):
        if val%7==0 and val%5!=0:
            lst.append(val)
    return lst

# Create a Python program that identifies all numbers between 100 and 300 (inclusive)
print(find_numbers(100, 200))
```

# Q2. Compute Factorial of a Number

```python
def factorial(num):
    product = 1
    for val in range(1,num+1):
        product *= val
    return product
# Create a function that calculates the factorial of a given number. The result should be returned as an integer.
print(factorial(8))
```

# Q3. Generate Square Dictionary

```python
def generate_square_dict(num):
    d = dict()
    for val in range(1,num+1):
        d[val] = val*val
    return d

# Create a Python function that takes an integer ( n ) as input and generates a dictionary containing pairs ( (i, i^2) ) # for all integers ( i ) from 1 to ( n ) (inclusive). The function should then return this dictionary.
print(generate_square_dict(8))
```

# Q6. Calculate Q Values from D

```python
# Create a Python function that computes the value of ( Q ) using the formula:
# where ( C ) is 50 and ( H ) is 30. The function should take only ( D ) as input, which consists of a comma-separated 
# sequence of values. The output should be rounded to the nearest integer and printed in a single line, separated by 
# commas.
import math as m
def calculate_q_values(values_string):
    l = values_string.split(",")
    q = []
    for d_val in l:
        d = int(d_val)
        c=50
        h=30
        q_val = m.sqrt((2*c*d)/h)
        q.append(m.floor(q_val))
    return q
print(calculate_q_values("100,150,180"))  # 100,150,180 are possible values of D
```

# Q7: Table Matrix

```python
# Create a Python program that takes two digits, M and N, as inputs and generates a two-dimensional array. 
# The value at the i-th row and j-th column of the array should be i*j.

def create_matrix(m, n):
    r = []
    for i in range(m):
        row = []
        for j in range(n):
            row.append(i*j)
        r.append(row)
    return r
    

print(create_matrix(4, 3))
```

# Q8. Sort Comma-Separated Words

```python
# Create a Python program that accepts a sequence of comma-separated words as input and returns the words sorted in 
# alphabetical order.
def sort_words(txt):
    l = txt.split(",")
    l.sort()
    
    return ",".join(l)
print(sort_words('banana,apple,grape,orange'))
```

# Q9. Capitalize All Lines

```python
# Create a Python program that takes a sequence of lines as input and  
# returns each line capitalized.


def capitalize_lines(list_lines):
    l = []
    for val in list_lines:
        l.append(val.upper())
    return l

print(capitalize_lines(['Sundays are Fun', 'Monday is Done']))
```

# Q10. Unique Sorted Words

```python
# Create a Python program that accepts a sequence of whitespace-separated words as input and returns them sorted alphabetically with duplicates removed.

def unique_sorted_words(txt):
    l = []
    for var in txt.split(" "):
        if var not in l:
            l.append(var)
    l.sort()
    return " ".join(l)
        
print(unique_sorted_words('dog cat apple cat banana dog'))
```

# Q11. Binary Divisible by 5

```python
# Create a Python program that accepts a sequence of comma-separated 4-digit binary numbers as input and checks if they are divisible by 5. The valid numbers should be returned in a comma-separated sequence.
def binary_divisible_by_5(txt):
    l = []
    for val in txt.split(","):
        num = int(val,2)
        if num%5==0:
            l.append(val)
    return ",".join(l)
            
        
print(binary_divisible_by_5('1101,1010,1111,1001'))
```

# Q12. All Even Digits

```python
# Create a Python program that finds all numbers between 1200 and 2100 (both inclusive) where each digit is an even number. The numbers should be returned as a comma-separated sequence.
def even_digit_numbers(m, n):
    l = []
    for num in range(m,n+1):
        flag=True
        for digit in str(num):
            if int(digit) %2!=0:
                flag=False
                break
        if flag==True:
            l.append(num)
    return l

print(even_digit_numbers(1200, 2010))
```

# Q13. Count Letters and Digits

```python
# Create a Python program that accepts a sentence and calculates the number of letters and digits.

def count_letters_digits(txt):
    num = 0
    letters=0
    for ch in txt:
        if ch.isdigit():
            num = num+1
        if ch.isalpha():
            letters=letters+1
    return { 'LETTERS':letters, 'DIGITS':num}
print(count_letters_digits('Data123 Science 2024'))
```

# Q14. Count Upper and Lower Case

```python
# Create a Python program that accepts a sentence and calculates the number of uppercase and lowercase letters.
def count_case(txt):
    lower_case_cnt = 0
    upper_case_cnt = 0
    for ch in txt:
        if ch >= 'A' and ch <='Z':
            upper_case_cnt +=1
        if ch >= 'a' and ch <='z':
            lower_case_cnt +=1
    return ({'UPPER CASE':upper_case_cnt , 'LOWER CASE':lower_case_cnt})
print(count_case('Hello World!'))
```

# Q15. Sum of a + aa + aaa + aaaa

```python
# Create a Python program that calculates the value of a + aa + aaa + aaaa for a given digit a.
def sum_of_series(target):
    s = 0
    i = 1
    while(i<=4):
        txt = str(target)*i
        s=s+int(txt)
        i = i+1
    return s
    
print(sum_of_series(7))
```

# Q16. Circle Class

```python
# Create a Python class Circle that accepts a radius as a parameter and has a method to compute the area of the circle.

class circle:
    def __init__(self,r):
        self.r = r
    def area(self):
        return 3.14*self.r*self.r
    
c=circle(5)
print(c.area())
```

# Q17. Shape and Square Area

```python
# Create a Python class Shape and a subclass Square. The Square class takes a length as a parameter and both classes have a method to compute the area. The area of Shape is 0 by default, while the area of Square is the square of its side length.

class shape:
    def __init__(self):
        pass
    def area(self):
        return 0
class square(shape):
    def __init__(self,l):
        self.l = l
    def area(self):
        return(self.l**2)
    
s = square(10)
print(s.area())
```

# Q20. Recursive Function for f(n)

```python
# Create a Python function f(n) such that f(n) = f(n-1) + 100 when n > 0 and f(0) = 1.
def f(n):
    if n==0:
        return 1
    else:
        return ( f(n-1)+100)
print(f(5))
```

# Q21. Generate Random Numbers

```python
# Write a Python function that generates a list of 5 random numbers between 150 and 250 (inclusive).
import random
def generate_random_numbers(m,n):
    l = []
    for i in range(1,6):
        l.append(random.randint(m,n))
    
    print(random.sample(range(m,n+1),5))
    return l
print(generate_random_numbers(150,250))
```

# Q22. Generate Even Numbers


```python
# Write a Python function that generates a list of 5 even numbers randomly between 150 and 250 (inclusive).
def generate_even_numbers(m,n):
    l = [i for i in range(m,n+1) if i%2==0]
    return(random.sample(l,5))

print(generate_even_numbers(150,250))
```

```python
# Write a Python program to measure the execution time of the expression 2+2 repeated 10000 times.
import time
def measure_execution_time():
    s = time.time()
    for _ in range(10000):
        2+2
    e = time.time()
    return f"Execution Time: {e - s} seconds"
print(measure_execution_time())
```
